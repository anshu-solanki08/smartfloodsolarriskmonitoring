```python
import joblib

# Load flood model
flood_model = joblib.load("../model/flood_model.pkl")
label_encoder = joblib.load("../model/label_encoder.pkl")

# Load solar model
solar_model = joblib.load("../model/solar_model.pkl")
flood_encoder = joblib.load("../model/flood_encoder.pkl")
damage_encoder = joblib.load("../model/damage_encoder.pkl")

# INPUT DATA
rainfall = 220
temperature = 26
humidity = 85
water_level = 8
soil_moisture = 80

cloud_cover = 75
waterlogging = 9

# Flood prediction
flood_input = [[rainfall, temperature, humidity, water_level, soil_moisture]]

flood_pred = flood_model.predict(flood_input)
flood_risk = label_encoder.inverse_transform(flood_pred)[0]

print("Flood Risk:", flood_risk)

# Solar prediction
encoded_flood = flood_encoder.transform([flood_risk])[0]

solar_input = [[encoded_flood, cloud_cover, humidity, waterlogging]]

solar_pred = solar_model.predict(solar_input)
solar_risk = damage_encoder.inverse_transform(solar_pred)[0]

print("Solar Damage Risk:", solar_risk)

# Efficiency logic
if solar_risk == "Low":
    efficiency = 90
elif solar_risk == "Medium":
    efficiency = 65
else:
    efficiency = 35

print("Solar Efficiency:", efficiency, "%")
```
